```python
import pandas as pd
import numpy as np 

from pathlib import Path 
import os 
import yaml

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
```

```python
os.chdir("..")
print(os.getcwd())
```

```python
data_path = Path("data/raw/student_academic_data/data.csv")
config_file_path = Path("config/config.yaml")

if not data_path.exists():
    raise FileNotFoundError

if not config_file_path.exists():
    raise FileNotFoundError
```

```python
with open(config_file_path, "r") as f:
    config = yaml.safe_load(f)

print(config)
```

```python
print(config["preprocessing"]["redundant_features"])
```

```python
print(config["preprocessing"]["test_size"])
```

```python
df = pd.read_csv(data_path, sep=";")
```

```python
df.columns = df.columns.str.strip().str.replace(r"\s+", " ", regex=True)
target = "Target"
```

```python
redundant_feature = config["preprocessing"]["redundant_features"]
test_size = config["preprocessing"]["test_size"]

print("--- redundant feature ---")
print(redundant_feature)

print(" --- test size ---")
print(test_size)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    df.drop(columns=[target]),
    df[target],
    test_size=test_size,
    random_state=42
)
```

```python
print(f"X_train, y_train shape: {X_train.shape}, {y_train.shape}")
```

```python
# step 1: remove redundant feature
print(f"before removing columns df size: {df.shape[1]}")

try:
    df.drop(columns=redundant_feature, inplace=True)
except KeyError:
    raise KeyError

print(f"after removing useless feature: {df.shape[1]}")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
le.fit(y_train)
y_train = le.transform(y_train)
y_test = le.transform(y_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier

MODELS = {
    "LogisticRegression": LogisticRegression(max_iter=1000),
    "RandomForest": RandomForestClassifier(random_state=42),
    "ExtraTrees": ExtraTreesClassifier(random_state=42),
    "XGBoost": XGBClassifier(random_state=42),
    "LightGBM": LGBMClassifier(random_state=42),
    "CatBoost": CatBoostClassifier(verbose=0)
}
```

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import StratifiedKFold, cross_val_score

import wandb

for name, model in MODELS.items():

    pipeline = Pipeline([
        ("classifier", model)
    ])

    scores = cross_val_score(
        pipeline,
        X_train,
        y_train,
        cv=5,
        scoring="f1_macro"
    )

    wandb.init(
        project="student-success-experiment-01",
        name=f"baseline-{name}",
        reinit=True
    )

    wandb.log({
        "model": name,
        "cv_mean": scores.mean(),
        "cv_std": scores.std()
    })

    wandb.finish()
```
